```python
from pylab import *
from matplotlib import font_manager as fm

def plot_pie(fracs, labels, atitle):
    # make a square figure and axes
    figure(1, figsize=(10,10))
    ax = axes([0.1, 0.1, 0.8, 0.8])

    explode=(0, 0.05, 0, 0.05, 0, 0.05, 0, 0.05)
    colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'plum', 'cyan', 'magenta', 'cadetblue']
    patches, texts, autotexts = pie(fracs, explode=explode, labels=labels, colors=colors,
                    autopct='%1.1f%%', shadow=True, startangle=90)

    title(atitle, fontsize = 20)
    
    proptease = fm.FontProperties()
    proptease.set_size('small')
    plt.setp(autotexts, fontproperties=proptease)
    plt.setp(texts, fontproperties=proptease)

    show()
```

```python
def make_100(fracs):
    fracs[7] = 100 - sum(fracs[:7])
    return fracs
```
